```python
import os
import cv2
from ultralytics import YOLO
import numpy as np

def load_model(model_path):
    if not os.path.exists(model_path):
        print(f"Error: Model file '{model_path}' not found.")
        exit()
    return YOLO(model_path)

def load_video(input_video_path):
    if not os.path.exists(input_video_path):
        print(f"Error: Video file '{input_video_path}' not found.")
        exit()
    cap = cv2.VideoCapture(input_video_path)
    if not cap.isOpened():
        print(f"Error: Could not open video file '{input_video_path}'.")
        exit()
    return cap

def get_video_properties(cap):
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    return frame_width, frame_height, fps, fourcc

def process_frame(frame, model):
    results = model(frame)
    for result in results:
        boxes = result.boxes.xyxy.numpy()
        confidences = result.boxes.conf.numpy()
        classes = result.boxes.cls.numpy()
        for box, conf, cls in zip(boxes, confidences, classes):
            x1, y1, x2, y2 = map(int, box)
            label = f"Pothole: {conf:.2f}"
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)
    return frame

def main():
    model_path = "best (5).pt"
    input_video_path = "kadri2.mp4"
    output_video_path = "detected_output.mp4"

    model = load_model(model_path)
    cap = load_video(input_video_path)
    frame_width, frame_height, fps, fourcc = get_video_properties(cap)

    out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        if frame is None:
            print("Error: Frame is empty.")
            continue

        processed_frame = process_frame(frame.copy(), model) # Create a copy of the frame
        out.write(processed_frame)

    cap.release()
    out.release()

    print(f"Output video saved to {output_video_path}")

if __name__ == "__main__":
    main()
```
